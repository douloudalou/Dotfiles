## Gui Element
- label - Widget to display text on the screen

```python
# import tkinter as library, tk is variable name storing the tkinter library
import tkinter as tk

# create window Tk() is create the window function
window = tk.Tk()

# create label
label = tk.Label(text="Hello world")
# pack will put the widgets into window, try run your code with or without pack
label.pack()

# window mainloop have to be called once at the end
# main loop process event
window.mainloop()
```

```python
# import tkinter as library
import tkinter as tk

# create window Tk() is create the window function
window = tk.Tk()

# create label with different parameters
label = tk.Label(
    text="Hello world",
    font=("Calibri",50), # Change the font size
    foreground="white",  # Set the text color to white
    background="black"  # Set the background color to black
)
# pack will put the widgets into window
label.pack()

# window mainloop have to be called once at the end
# main loop process event
window.mainloop()
```

## Gui Element
- Button - contain text and perform action when clicked

https://tkdocs.com/pyref/ttk_button.html

```python
# import tkinter as library
import tkinter as tk

# create window Tk() is create the window function
window = tk.Tk()

#create a button with different parameter
button = tk.Button(
    text="Button",
    bg="green",
    fg="white"
)
button.pack()

# window mainloop have to be called once at the end
# main loop process event
window.mainloop()
```

```python
# import tkinter as library
import tkinter as tk

# create window Tk() is create the window function
window = tk.Tk()

#An example of 3 different button, with different parameters

#create a button with different parameter
button = tk.Button(
    text="Button",
    bg="green",
    fg="white"
)
button.pack()

#Notice the button length will change according to text size
longer_button = tk.Button(
    text="Longer Button",
    bg="green",
    fg="white"
)
longer_button.pack()

# button with width and height
width_button = tk.Button(
    text="width 20 Button",
    bg="green",
    fg="white",
    width = 20,
    height = 1
)
width_button.pack()


# window mainloop have to be called once at the end
# main loop process event
window.mainloop()
```

## Gui Element
- Frame - a rectangular region used to contain related widget

https://tkdocs.com/pyref/ttk_frame.html

```python
# Creating a simple Frame 
import tkinter as tk

window = tk.Tk()

# create a frame 
frame = tk.Frame()
# add the frame to the window
frame.pack()

#there is nothing in the frame so the window is empty
window.mainloop()
```

```python
# Creating a frame that contains button

import tkinter as tk

window = tk.Tk()

#create  frame
frame= tk.Frame()
 
# putting button into frame by setting the master variable as frame
button = tk.Button(master=frame, text="Button in frame")
button.pack()

# Try not to pack the frame and see what will happen
frame.pack()

window.mainloop()
```

```python
# Create a frame that contain button. A label without a frame
import tkinter as tk

window = tk.Tk()

#create  frame
frame= tk.Frame()
 
# putting button into frame by setting the master variable as frame
button = tk.Button(master=frame, text="Button in frame")
button.pack()

label = tk.Label(text="Label not in Frame")
# pack will put the widgets into window
label.pack()

frame.pack()
# Why does the button appear after the label eventhough it is packed earlier?
# ans: As the button is pack into the frame and the frame is pack after the label

window.mainloop()
```

## Geometry Managers

The layout of the application are controlled by geometry managers. We have been using pack() to pack widget to the windows.

Another 2 options:
- .place()
- .grid()

## pack ()

When a new widget are added using pack(), it will be added to the window that is just tall and wide enough to hold the widget. The widget will be center.

## place()
When a new widget are added using place(x,y), it will be added based on the x and y. Position(0,0) is top left of the window

```python
import tkinter as tk

window = tk.Tk()

# putting label at position 50, 50
label_a = tk.Label(text="(50,50)")
# using 
label_a.place(x=50,y=50)

# putting label at position 100, 100
label_b = tk.Label(text="(100,100)")
# using 
label_b.place(x=100,y=100)

window.mainloop()
```

## grid()
grid() work by dividing the window into row and column.

```python
import tkinter as tk

window = tk.Tk()

# Create a frame at row 0 and column 0
frame = tk.Frame(
        master=window,
        relief=tk.RAISED,
        borderwidth=1)
# frame position in row and column
frame.grid(row=0, column=0, padx=5, pady=5)
#label put in frame
label = tk.Label(master=frame, text=f"Row 0\nColumn 0")
# label pack into window
label.pack()

# Create a frame at row 0 and column 1
frame1 = tk.Frame(
        master=window,
        relief=tk.RAISED, # change the look of the frame to be raised
        borderwidth=1)
# frame position in row and column
frame1.grid(row=0, column=1, padx=5, pady=5)
#label put in frame
label = tk.Label(master=frame1, text=f"Row 0\nColumn 1")
# label pack into window
label.pack()

# Create a frame at row 0 and column 2
frame2 = tk.Frame(
        master=window,
        relief=tk.RAISED,
        borderwidth=1)
# frame position in row and column
frame2.grid(row=0, column=2, padx=5, pady=5)
#label put in frame
label = tk.Label(master=frame2, text=f"Row 0\nColumn 2")
# label pack into window
label.pack()

window.mainloop()
```

## Events
- an action that happen which might trigger some behavior in the application, such as key pressed or button clicked.

## Handling button event

```python
import tkinter as tk

# Create a window object
window = tk.Tk()

def handle_click(event):
    print("The button was clicked!")

button = tk.Button(text="Button!")

button.bind("<Button-1>", handle_click)

button.pack()

# Run the event loop
window.mainloop()
```
